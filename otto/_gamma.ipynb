```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
tau = 0.01
m = 0.005
T = 2000

print("Expected time to break: ", 1 / tau)
print("Expected time to repair: ", 1 / m)

t = np.linspace(0, T, 100000)
```

```python
def pdf_exp(t, scale: float):
    return scale * np.exp(-scale * t)

plt.plot(t, pdf_exp(t, tau), label="time to break")
plt.plot(t, pdf_exp(t, m), label="time to repair")
plt.legend()
plt.xlabel("seconds")
plt.show()
```

```python
shape, scale = 2, tau  # mean=4, std=2*sqrt(2)
s = np.random.gamma(shape, scale, 100000)
count, bins, ignored = plt.hist(s, 50, density=True, )
```
